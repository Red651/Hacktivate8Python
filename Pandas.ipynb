```python
further_city_data = pd.DataFrame({'revenue': [7000, 3400],
                                'employee_count': [2, 2]},
                                index=['New York', 'Barcelona'])
further_city_data        
```

```python
city_data
```

```python
all_city_data = pd.concat([city_data, further_city_data]) # untuk menggabungkan data
```

```python
all_city_data
```

```python
city_coutry = pd.DataFrame({'country': ['Holland', 'Holland', 'Japan', 'USA', 'USA'],
                            'capital': [1, 0, 0, 1, 0]},
                            index=['Amsterdam', 'Rotterdam', 'Tokyo', 'New York', 'Barcelona'])
```

```python
city_coutry
```

```python
cities = pd.concat([all_city_data, city_coutry], axis=1, sort=False) # untuk menggabungkan data
cities
```

```python
pd.concat([all_city_data, city_coutry], axis=1, sort=False, join='inner')
# inner untuk menggabungkan data yang sama saja
```

```python
countries = pd.DataFrame({'population_millions': [17, 127, 37],
                        'continent': ['Europe', 'North America', 'Asia']},
                        index=['Holland', 'USA', 'Japan'])
```

```python
countries
```

```python
cities
```

```python
pd.merge(cities, countries, left_on='country', right_index=True) # untuk menggabungkan data
```
